```python
#4월 지하철 이용승객정보
df = pd.read_excel('/content/drive/MyDrive/subway.xlsx', thousands = ',')
df.head()
```

```python
df.tail()
```

```python
df1=df.groupby('노선명').count()
df1
```

```python
plt.rc('font', family='NanumBarunGothic') 
plt.rcParams["figure.figsize"] = (15,5)

plt.bar(df1.index, df1['역명'], color='red')

plt.title('지하철 역 현황', fontsize=20)
plt.xlabel('역이름', fontsize=14)
plt.ylabel('개수',rotation='horizontal',fontsize=14)
```

```python
df1=df.sort_values(by='승차총승객수', ascending=False)
df1
```

```python
#승차총객수가 가장 많은 역의 인덱스
idx=df['승차총승객수'].idxmax(axis=0, skipna=True)
idx
```

```python
#승차총객수가 가장 많은 역의 역명
df.loc[idx,'역명' ]
```

```python
#승차총객수가 가장 작은 역의 인덱스
idx=df['승차총승객수'].idxmin(axis=0, skipna=True)
idx
```

```python
#승차총객수가 가장 작은 역의 역명
df.loc[idx,'역명' ]
```
