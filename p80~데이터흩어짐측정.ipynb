```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/2022_DataScience/seoul_temp.xlsx', thousands = ',')
df.head()
```

```python
df.tail()
```

```python
## 문자열을 Datetime으로 변환하기
df['날짜'] = pd.to_datetime(df['날짜'],format='%Y-%m-%d')
df
```

```python
df1=df[(df['날짜']>='2021-05-01') & (df['날짜']<='2021-05-31')]
df1
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(1,32), df1['평균기온(℃)'], marker=[5,1])
plt.title('Average Temp')
plt.xlabel('2021-05-01~31')
plt.ylabel('temp')
```
